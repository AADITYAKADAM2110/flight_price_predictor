```python
from category_encoders import TargetEncoder

encoder = TargetEncoder()
dataset_without_flight_no['airline'] = encoder.fit_transform(dataset_without_flight_no['airline'], dataset_without_flight_no['price'])
```

```python
dataset_without_flight_no
```

```python
dataset_without_flight_no['source_city'] = encoder.fit_transform(dataset_without_flight_no['source_city'], dataset_without_flight_no['price'])
dataset_without_flight_no['destination_city'] = encoder.fit_transform(dataset_without_flight_no['destination_city'], dataset_without_flight_no['price'])
```

```python
dataset_without_flight_no
```

```python
dataset_without_flight_no.info()
```

```python
dataset_without_flight_no.drop(['flight'], axis=1, inplace=True)

# dropped the 'flight' column as it is not needed for prediction
```

```python
dataset_without_flight_no
```

```python
dataset_without_flight_no['stops'] = dataset_without_flight_no['stops'].map({'zero': 0, 'one': 1, 'two_or_more': 2})
dataset_without_flight_no['class'] = dataset_without_flight_no['class'].map({'Economy': 0, 'Business': 1})
dataset_without_flight_no['departure_time'] = encoder.fit_transform(dataset_without_flight_no['departure_time'], dataset_without_flight_no['price'])
dataset_without_flight_no['arrival_time'] = encoder.fit_transform(dataset_without_flight_no['arrival_time'], dataset_without_flight_no['price'])
```

```python
dataset_without_flight_no.info()
```

```python
dataset_without_flight_no.corr()['price'].sort_values(ascending=False)
```

```python
dataset_without_flight_no.to_csv('Preprocessed_flight_data.csv', index=False) 
# saved the preprocessed dataset to a CSV file
print("Preprocessed dataset saved successfully.")
```

```python
dataset_for_feature_engineering = pd.read_csv(r'C:\Users\DELL\Desktop\Flight Ticket Prediction Project\dataset\copy_of_preprocessed_flight_data.csv')
```

```python
dataset_for_feature_engineering
```

```python
dataset_for_feature_engineering['airline_class'] = dataset_for_feature_engineering['airline'].astype(str) + '_' + dataset_for_feature_engineering['class'].astype(str)
dataset_for_feature_engineering['duration_stops'] = dataset_for_feature_engineering['duration'].astype(str) + '_' + dataset_for_feature_engineering['stops'].astype(str)
```

```python
dataset_for_feature_engineering
```

```python
dataset_for_feature_engineering.info()
```

```python
dataset_for_feature_engineering['airline_class'] = encoder.fit_transform(dataset_for_feature_engineering['airline_class'], dataset_for_feature_engineering['price'])
dataset_for_feature_engineering['duration_stops'] = encoder.fit_transform(dataset_for_feature_engineering['duration_stops'], dataset_for_feature_engineering['price'])
```

```python
dataset_for_feature_engineering
```

```python
dataset_for_feature_engineering.corr()['price'].sort_values(ascending=False)
```

```python
dataset_for_feature_engineering['is_direct_flight'] = dataset_for_feature_engineering['stops'].apply(lambda x: 1 if x == 0 else 0)
```

```python
dataset_for_feature_engineering['avg_airline_price'] = dataset_for_feature_engineering.groupby('airline')['price'].transform('mean')
```

```python
dataset_for_feature_engineering.corr()['price'].sort_values(ascending=False)
```

```python
dataset_for_feature_engineering.drop(['airline', 'class', 'duration', 'stops'], axis=1, inplace=True)
```

```python
dataset_for_feature_engineering
```

```python
dataset_for_feature_engineering.to_csv('Feature_engineered_flight_data.csv', index=False)
# saved the feature engineered dataset to a CSV file
print("Feature engineered dataset saved successfully.")
```
