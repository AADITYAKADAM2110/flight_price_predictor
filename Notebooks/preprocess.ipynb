```python
from category_encoders import TargetEncoder

encoder = TargetEncoder()
dataset_without_flight_no['airline'] = encoder.fit_transform(dataset_without_flight_no['airline'], dataset_without_flight_no['price'])
```

```python
dataset_without_flight_no
```

```python
dataset_without_flight_no['source_city'] = encoder.fit_transform(dataset_without_flight_no['source_city'], dataset_without_flight_no['price'])
dataset_without_flight_no['destination_city'] = encoder.fit_transform(dataset_without_flight_no['destination_city'], dataset_without_flight_no['price'])
```

```python
dataset_without_flight_no
```

```python
dataset_without_flight_no.info()
```

```python
dataset_without_flight_no.drop(['flight'], axis=1, inplace=True)

# dropped the 'flight' column as it is not needed for prediction
```

```python
dataset_without_flight_no
```

```python
dataset_without_flight_no['stops'] = dataset_without_flight_no['stops'].map({'zero': 0, 'one': 1, 'two_or_more': 2})
dataset_without_flight_no['class'] = dataset_without_flight_no['class'].map({'Economy': 0, 'Business': 1})
dataset_without_flight_no['departure_time'] = encoder.fit_transform(dataset_without_flight_no['departure_time'], dataset_without_flight_no['price'])
dataset_without_flight_no['arrival_time'] = encoder.fit_transform(dataset_without_flight_no['arrival_time'], dataset_without_flight_no['price'])
```
